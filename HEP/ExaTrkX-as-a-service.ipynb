# ExaTrkX as a Service Performance

Results for CTD 2023.

```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
cpu_results = [(1, 924.2), (25, 608.8), 
               (50, 434.22), (100, 432.18), (256, 462.83), (1000, 431.20)]
cpu_triton_custom = [100, 70.3]
```

```python
threads = np.array([x[0] for x in cpu_results[:4]])
latency = np.array([x[1] for x in cpu_results[:4]])
```

```python
fig, ax = plt.subplots(1, 1, figsize=(8, 6))
plt.plot(threads, latency, "-o", c="k", lw=2, markersize=10, label="Direct Inference TBB")
plt.plot([cpu_triton_custom[0]], [cpu_triton_custom[1]], "-o", c="r", lw=2, markersize=10, label="Triton")
plt.xlabel("# of threads", fontsize=16)
plt.ylabel("Total time for 100 events [s]", fontsize=16)
plt.legend(fontsize=16)
plt.title("One Perlmutter CPU node")
plt.grid()
fig.tight_layout()
plt.savefig("triton_cpu.png")
plt.show()
```
