2HDM couplings:
* $\kappa$(h, V) = $\sin(\beta - \alpha)$, $\kappa$(H, V) = $\cos(\beta - \alpha)$
* $\kappa$(H, t) = -$\sin(\beta - \alpha)/\tan\beta + \cos(\beta - \alpha) = \sin\alpha/\sin\beta$, 
* Type-I: $\kappa$(H, b) = $\kappa$(H, t)
* Type-II: $\kappa$(H, b) = $\sin(\beta - \alpha)\times\tan\beta + \cos(\beta - \alpha) = \cos\alpha/\cos\beta$

$\beta$ in [0, $\pi$/2] and ($\beta - \alpha$) in [0, $\pi$].

```python
import numpy as np
```

```python
x = np.linspace(-1, 1, 100) # tanb
y = np.linspace(1, 10, 100) # cos(b-a)
```

```python
import itertools
```

```python
comb = np.array(list(itertools.product(x, y)))
x0, y0 = comb[:, 0], comb[:, 1]
```

```python
def h_top(x, y):
    z = np.sqrt(1-x**2)
    return -z/y + x
```

```python
def b_bottom(x, y):
    z = np.sqrt(1-x**2)
    return z*y + x
```

```python
top_coupling = h_top(x0, y0)
bottom_coupling = b_bottom(x0, y0)
```

```python
import matplotlib.pyplot as plt
```

```python
import pandas as pd
```

```python
df = pd.DataFrame(comb, columns=['x', 'y'])
df = df.assign(t=top_coupling, b=bottom_coupling)
```

```python
fig, ax = plt.subplots(figsize=(8, 6))
df.plot.scatter(x='x', y='y', c='t', s=50, cmap='bwr', ax=ax, vmin=-1, vmax=1)
plt.savefig('2hdm-top-coupling.pdf')
```

```python
fig, ax = plt.subplots(figsize=(8, 6))
df.plot.scatter(x='x', y='y', c='b', s=50, cmap='Blues', ax=ax, vmin=-1)
```
